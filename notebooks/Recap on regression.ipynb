```python
# Test out some data
import plotly
from plotly.offline import init_notebook_mode
init_notebook_mode(connected = True)

import plotly.express as px
from src.data.load import get_data


data_id = "solar"
df = get_data(data_id)
df.reset_index(inplace=True)

fig = px.scatter(data_frame=df, x="x", y="y")
plotly.offline.iplot(fig)
```

```python
import tensorflow as tf
from src.models.linear import LinearRegression


tf.config.set_visible_devices([], "GPU")  # My M1 acting up
linear_model = LinearRegression()
linear_model.fit(df["x"].values, df["y"].values)
preds = linear_model.predict(df["x"].values)
```

```python
import plotly.graph_objects as go

fig.add_traces(go.Scatter(x=df["x"], y=preds, mode="lines", name="predictions"))
```

```python
from sklearn.metrics import mean_squared_error

print(f"Mean error: {mean_squared_error(preds, df['y'])}")
```
